```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# All

Onset_Phonemes_All = []
Coda_Phonemes_All = []
Classes_All = []

path_audio = '../../data/external/AVP_Dataset/Personal'

list_wav_all = []
list_csv_all = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('.wav'):
            list_wav_all.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv_all.append(os.path.join(path, filename))

list_wav_all = sorted(list_wav_all)
list_csv_all = sorted(list_csv_all)

list_wav_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))
list_csv_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))

for i in range(len(list_csv_all)):
    
    Classes = pd.read_csv(list_csv_all[i], header=None)[1].tolist()
    Onset_Phonemes = pd.read_csv(list_csv_all[i], header=None)[2].tolist()
    Coda_Phonemes = pd.read_csv(list_csv_all[i], header=None)[3].tolist()
    
    for n in range(len(Classes)):
        Classes_All.append(Classes[n])
        Onset_Phonemes_All.append(Onset_Phonemes[n])
        Coda_Phonemes_All.append(Coda_Phonemes[n])

path_audio = '../../data/external/LVT_Dataset/DataSet_Wav_Annotation'

list_wav = []
list_csv = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('1.wav'):
            list_wav.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv.append(os.path.join(path, filename))

list_wav = sorted(list_wav)
list_csv = sorted(list_csv)

for i in range(len(list_csv)):
    
    Classes = pd.read_csv(list_csv[i], header=None)[1].tolist()
    Onset_Phonemes = pd.read_csv(list_csv[i], header=None)[2].tolist()
    Coda_Phonemes = pd.read_csv(list_csv[i], header=None)[3].tolist()
    
    for n in range(len(Classes)):
        Classes_All.append(Classes[n])
        Onset_Phonemes_All.append(Onset_Phonemes[n])
        Coda_Phonemes_All.append(Coda_Phonemes[n])
```

```python
# Train-Val

Onset_Phonemes_TrainVal = []
Coda_Phonemes_TrainVal = []
Classes_TrainVal = []

path_audio = '../../data/external/AVP_Dataset/Personal'

list_wav_all = []
list_csv_all = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('.wav'):
            list_wav_all.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv_all.append(os.path.join(path, filename))

list_wav_all = sorted(list_wav_all)
list_csv_all = sorted(list_csv_all)

list_wav_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))
list_csv_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))

for i in range(len(list_csv_all)):

    if 'Participant_9' in list_csv_all[i] or 'Participant_11' in list_csv_all[i] or 'Participant_19' in list_csv_all[i] or 'Participant_24' in list_csv_all[i]:
        continue
    
    Classes = pd.read_csv(list_csv_all[i], header=None)[1].tolist()
    Onset_Phonemes = pd.read_csv(list_csv_all[i], header=None)[2].tolist()
    Coda_Phonemes = pd.read_csv(list_csv_all[i], header=None)[3].tolist()
    
    for n in range(len(Classes)):
        Classes_TrainVal.append(Classes[n])
        Onset_Phonemes_TrainVal.append(Onset_Phonemes[n])
        Coda_Phonemes_TrainVal.append(Coda_Phonemes[n])

path_audio = '../../data/external/LVT_Dataset/DataSet_Wav_Annotation'

list_wav = []
list_csv = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('1.wav'):
            list_wav.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv.append(os.path.join(path, filename))

list_wav = sorted(list_wav)
list_csv = sorted(list_csv)

for i in range(len(list_csv)):

    if 'Participant_1' in list_csv_all[i] or 'Participant_7' in list_csv_all[i] or 'Participant_8' in list_csv_all[i] or 'Participant_14' in list_csv_all[i]:
        continue
    
    Classes = pd.read_csv(list_csv[i], header=None)[1].tolist()
    Onset_Phonemes = pd.read_csv(list_csv[i], header=None)[2].tolist()
    Coda_Phonemes = pd.read_csv(list_csv[i], header=None)[3].tolist()
    
    for n in range(len(Classes)):
        Classes_TrainVal.append(Classes[n])
        Onset_Phonemes_TrainVal.append(Onset_Phonemes[n])
        Coda_Phonemes_TrainVal.append(Coda_Phonemes[n])
```

```python
# Evaluation

Onset_Phonemes_Eval = []
Coda_Phonemes_Eval = []
Classes_Eval = []

path_audio = '../../data/external/AVP_Dataset/Personal'

list_wav_all = []
list_csv_all = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('.wav'):
            list_wav_all.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv_all.append(os.path.join(path, filename))

list_wav_all = sorted(list_wav_all)
list_csv_all = sorted(list_csv_all)

list_wav_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))
list_csv_all.sort(key = lambda f:int(''.join(filter(str.isdigit,f))))

for i in range(len(list_csv_all)):

    if 'Participant_9' in list_csv_all[i] or 'Participant_11' in list_csv_all[i] or 'Participant_19' in list_csv_all[i] or 'Participant_24' in list_csv_all[i]:
        Classes = pd.read_csv(list_csv_all[i], header=None)[1].tolist()
        Onset_Phonemes = pd.read_csv(list_csv_all[i], header=None)[2].tolist()
        Coda_Phonemes = pd.read_csv(list_csv_all[i], header=None)[3].tolist()
        for n in range(len(Classes)):
            Classes_Eval.append(Classes[n])
            Onset_Phonemes_Eval.append(Onset_Phonemes[n])
            Coda_Phonemes_Eval.append(Coda_Phonemes[n])
    else:
        continue

path_audio = '../../data/external/LVT_Dataset/DataSet_Wav_Annotation'

list_wav = []
list_csv = []

for path, subdirs, files in os.walk(path_audio):
    for filename in files:
        if filename.endswith('1.wav'):
            list_wav.append(os.path.join(path, filename))
        if filename.endswith('.csv'):
            list_csv.append(os.path.join(path, filename))

list_wav = sorted(list_wav)
list_csv = sorted(list_csv)

for i in range(len(list_csv)):
    
    if 'Participant_1' in list_csv_all[i] or 'Participant_7' in list_csv_all[i] or 'Participant_8' in list_csv_all[i] or 'Participant_14' in list_csv_all[i]:
        Classes = pd.read_csv(list_csv[i], header=None)[1].tolist()
        Onset_Phonemes = pd.read_csv(list_csv[i], header=None)[2].tolist()
        Coda_Phonemes = pd.read_csv(list_csv[i], header=None)[3].tolist()
        for n in range(len(Classes)):
            Classes_Eval.append(Classes[n])
            Onset_Phonemes_Eval.append(Onset_Phonemes[n])
            Coda_Phonemes_Eval.append(Coda_Phonemes[n])
    else:
        continue
```

```python
# Make syllables

Syllables_All = []
for n in range(len(Onset_Phonemes_All)):
    Syllables_All.append(Onset_Phonemes_All[n]+Coda_Phonemes_All[n])

Syllables_TrainVal = []
for n in range(len(Onset_Phonemes_TrainVal)):
    Syllables_TrainVal.append(Onset_Phonemes_TrainVal[n]+Coda_Phonemes_TrainVal[n])

Syllables_Eval = []
for n in range(len(Onset_Phonemes_Eval)):
    Syllables_Eval.append(Onset_Phonemes_Eval[n]+Coda_Phonemes_Eval[n])
```

```python
# Count All

Classes_All_Unique = sorted(np.unique(Classes_All))
Classes_All_Percentage = np.zeros(len(Classes_All_Unique)-3)
for n in range(3,len(Classes_All_Unique)):
    if n==3:
        count = Classes_All.count(Classes_All_Unique[n]) + Classes_All.count(Classes_All_Unique[0])
        Classes_All_Percentage[n-3] = np.round((count/len(Classes_All))*100, 1)
    elif n==5:
        count = Classes_All.count(Classes_All_Unique[n]) + Classes_All.count(Classes_All_Unique[1])
        Classes_All_Percentage[n-3] = np.round((count/len(Classes_All))*100, 1)
    elif n==6:
        count = Classes_All.count(Classes_All_Unique[n]) + Classes_All.count(Classes_All_Unique[2])
        Classes_All_Percentage[n-3] = np.round((count/len(Classes_All))*100, 1)
    else:
        count = Classes_All.count(Classes_All_Unique[n])
        Classes_All_Percentage[n-3] = np.round((count/len(Classes_All))*100, 1)
Classes_All_Unique = Classes_All_Unique[3:]

Onset_Phonemes_All_Unique = np.unique(Onset_Phonemes_All)
Onset_Phonemes_All_Percentage = np.zeros(len(Onset_Phonemes_All_Unique))
for n in range(len(Onset_Phonemes_All_Unique)):
    count = Onset_Phonemes_All.count(Onset_Phonemes_All_Unique[n])
    Onset_Phonemes_All_Percentage[n] = np.round((count/len(Onset_Phonemes_All))*100, 1)

Coda_Phonemes_All_Unique = np.unique(Coda_Phonemes_All)
Coda_Phonemes_All_Percentage = np.zeros(len(Coda_Phonemes_All_Unique))
for n in range(len(Coda_Phonemes_All_Unique)):
    count = Coda_Phonemes_All.count(Coda_Phonemes_All_Unique[n])
    Coda_Phonemes_All_Percentage[n] = np.round((count/len(Coda_Phonemes_All))*100, 1)

Syllables_All_Unique = np.unique(Syllables_All)
Syllables_All_Percentage = np.zeros(len(Syllables_All_Unique))
for n in range(len(Syllables_All_Unique)):
    count = Syllables_All.count(Syllables_All_Unique[n])
    Syllables_All_Percentage[n] = np.round((count/len(Syllables_All))*100, 1)
```

```python
# Count TrainVal

Classes_TrainVal_Unique = sorted(np.unique(Classes_TrainVal))
Classes_TrainVal_Percentage = np.zeros(len(Classes_TrainVal_Unique)-3)
for n in range(3,len(Classes_TrainVal_Unique)):
    if n==3:
        count = Classes_TrainVal.count(Classes_TrainVal_Unique[n]) + Classes_TrainVal.count(Classes_TrainVal_Unique[0])
        Classes_TrainVal_Percentage[n-3] = np.round((count/len(Classes_TrainVal))*100, 1)
    elif n==5:
        count = Classes_TrainVal.count(Classes_TrainVal_Unique[n]) + Classes_TrainVal.count(Classes_TrainVal_Unique[1])
        Classes_TrainVal_Percentage[n-3] = np.round((count/len(Classes_TrainVal))*100, 1)
    elif n==6:
        count = Classes_TrainVal.count(Classes_TrainVal_Unique[n]) + Classes_TrainVal.count(Classes_TrainVal_Unique[2])
        Classes_TrainVal_Percentage[n-3] = np.round((count/len(Classes_TrainVal))*100, 1)
    else:
        count = Classes_TrainVal.count(Classes_TrainVal_Unique[n])
        Classes_TrainVal_Percentage[n-3] = np.round((count/len(Classes_TrainVal))*100, 1)
Classes_TrainVal_Unique = Classes_TrainVal_Unique[3:]

Onset_Phonemes_TrainVal_Unique = np.unique(Onset_Phonemes_TrainVal)
Onset_Phonemes_TrainVal_Percentage = np.zeros(len(Onset_Phonemes_TrainVal_Unique))
for n in range(len(Onset_Phonemes_TrainVal_Unique)):
    count = Onset_Phonemes_TrainVal.count(Onset_Phonemes_TrainVal_Unique[n])
    Onset_Phonemes_TrainVal_Percentage[n] = np.round((count/len(Onset_Phonemes_TrainVal))*100, 1)

Coda_Phonemes_TrainVal_Unique = np.unique(Coda_Phonemes_TrainVal)
Coda_Phonemes_TrainVal_Percentage = np.zeros(len(Coda_Phonemes_TrainVal_Unique))
for n in range(len(Coda_Phonemes_TrainVal_Unique)):
    count = Coda_Phonemes_TrainVal.count(Coda_Phonemes_TrainVal_Unique[n])
    Coda_Phonemes_TrainVal_Percentage[n] = np.round((count/len(Coda_Phonemes_TrainVal))*100, 1)

Syllables_TrainVal_Unique = np.unique(Syllables_TrainVal)
Syllables_TrainVal_Percentage = np.zeros(len(Syllables_TrainVal_Unique))
for n in range(len(Syllables_TrainVal_Unique)):
    count = Syllables_TrainVal.count(Syllables_TrainVal_Unique[n])
    Syllables_TrainVal_Percentage[n] = np.round((count/len(Syllables_TrainVal))*100, 1)
```

```python
# Count Eval

Classes_Eval_Unique = sorted(np.unique(Classes_Eval))
Classes_Eval_Percentage = np.zeros(len(Classes_Eval_Unique)-3)
for n in range(3,len(Classes_Eval_Unique)):
    if n==3:
        count = Classes_Eval.count(Classes_Eval_Unique[n]) + Classes_Eval.count(Classes_Eval_Unique[0])
        Classes_Eval_Percentage[n-3] = np.round((count/len(Classes_Eval))*100, 1)
    elif n==5:
        count = Classes_Eval.count(Classes_Eval_Unique[n]) + Classes_Eval.count(Classes_Eval_Unique[1])
        Classes_Eval_Percentage[n-3] = np.round((count/len(Classes_Eval))*100, 1)
    elif n==6:
        count = Classes_Eval.count(Classes_Eval_Unique[n]) + Classes_Eval.count(Classes_Eval_Unique[2])
        Classes_Eval_Percentage[n-3] = np.round((count/len(Classes_Eval))*100, 1)
    else:
        count = Classes_Eval.count(Classes_Eval_Unique[n])
        Classes_Eval_Percentage[n-3] = np.round((count/len(Classes_Eval))*100, 1)
Classes_Eval_Unique = Classes_Eval_Unique[3:]

Onset_Phonemes_Eval_Unique = np.unique(Onset_Phonemes_Eval)
Onset_Phonemes_Eval_Percentage = np.zeros(len(Onset_Phonemes_Eval_Unique))
for n in range(len(Onset_Phonemes_Eval_Unique)):
    count = Onset_Phonemes_Eval.count(Onset_Phonemes_Eval_Unique[n])
    Onset_Phonemes_Eval_Percentage[n] = np.round((count/len(Onset_Phonemes_Eval))*100, 1)

Coda_Phonemes_Eval_Unique = np.unique(Coda_Phonemes_Eval)
Coda_Phonemes_Eval_Percentage = np.zeros(len(Coda_Phonemes_Eval_Unique))
for n in range(len(Coda_Phonemes_Eval_Unique)):
    count = Coda_Phonemes_Eval.count(Coda_Phonemes_Eval_Unique[n])
    Coda_Phonemes_Eval_Percentage[n] = np.round((count/len(Coda_Phonemes_Eval))*100, 1)

Syllables_Eval_Unique = np.unique(Syllables_Eval)
Syllables_Eval_Percentage = np.zeros(len(Syllables_Eval_Unique))
for n in range(len(Syllables_Eval_Unique)):
    count = Syllables_Eval.count(Syllables_Eval_Unique[n])
    Syllables_Eval_Percentage[n] = np.round((count/len(Syllables_Eval))*100, 1)
```

```python
df1=pd.DataFrame({'x':Syllables_All_Unique,'y':Syllables_All_Percentage})
df2=pd.DataFrame({'x':Syllables_TrainVal_Unique,'y':Syllables_TrainVal_Percentage})
df3=pd.DataFrame({'x':Syllables_Eval_Unique,'y':Syllables_Eval_Percentage})

df1['Sets'] = 'All'
df2['Sets'] = 'Train-Validation'
df3['Sets'] = 'Evaluation'

plt.figure(figsize=(36,8))
concatenated = pd.concat([df1,df2,df3])
sns.barplot(x='x', y='y', data=concatenated, hue='Sets')
plt.legend(loc='upper right', fontsize='xx-large')
plt.xlabel('Syllables', fontsize=20)
plt.ylabel('Percentage', fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```
